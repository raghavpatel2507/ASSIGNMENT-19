```python
#q-1
Estimation statistics involves estimating unknown population parameters based on sample data.

Point estimate: It is a single value that serves as the best guess for the population parameter. It is obtained by calculating a sample statistic, such as the sample mean or sample proportion. Point estimates provide a specific value but do not indicate the uncertainty associated with the estimate.

Interval estimate: Also known as a confidence interval, it provides a range of values within which the population parameter is likely to fall. It considers both the point estimate and the variability of the estimate. Interval estimates are accompanied by a confidence level, such as 95% or 99%, which indicates the probability that the true parameter lies within the interval. Interval estimates provide a measure of uncertainty and are often more informative than point estimates alone.
```

```python
#q-2
import math

def estimate_population_mean(sample_mean, sample_std_dev, sample_size):
    
    standard_error = sample_std_dev / math.sqrt(sample_size)
    z_score = 1.96  # Assuming a 95% confidence level (corresponding to a z-score of 1.96)
    margin_of_error = z_score * standard_error
    
    lower_limit = sample_mean - margin_of_error
    upper_limit = sample_mean + margin_of_error
    
    
    return lower_limit, upper_limit
```

```python
#q-5
import numpy as np
from scipy.stats import t

def hypothesis_test(sample1, sample2, alpha):
    
    n1 = len(sample1)
    n2 = len(sample2)
    
    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)
    
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)
    
    # Compute the pooled standard deviation
    pooled_std_dev = np.sqrt((var1/n1) + (var2/n2))
    
    # Compute the test statistic
    t_stat = (mean1 - mean2) / pooled_std_dev
    
    # Compute the degrees of freedom
    degrees_of_freedom = n1 + n2 - 2
    
    # Compute the critical value for a two-tailed test
    critical_value = t.ppf(1 - alpha/2, degrees_of_freedom)
    
    # Compare the test statistic with the critical value
    if np.abs(t_stat) > critical_value:
        return True  # Reject the null hypothesis
    else:
        return False  # Fail to reject the null hypothesis
```

```python
sample1 = [82, 85, 88, 90, 79, 92, 86, 88, 85, 90]
sample2 = [75, 78, 82, 80, 83, 77, 84, 79, 81, 85]
alpha = 0.05

result = hypothesis_test(sample1, sample2, alpha)

if result:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
#q-6
The null hypothesis (H₀) and alternative hypothesis (H₁) are statements formulated in hypothesis testing to assess the evidence and make conclusions about a population based on sample data.

The null hypothesis (H₀) represents the default or conservative assumption, stating that there is no significant effect, difference, or relationship in the population. It assumes that any observed differences or relationships in the sample are due to random chance or sampling variability. The null hypothesis is often denoted as no effect, no difference, or no association.

The alternative hypothesis (H₁), also known as the research hypothesis or experimental hypothesis, represents the claim or assertion that contradicts the null hypothesis. It states that there is a specific effect, difference, or relationship in the population that is of interest to the researcher. The alternative hypothesis can take different forms, depending on the research question or the nature of the study.

Example: Gender Differences
Null Hypothesis (H₀): There is no significant difference in mean salaries between male and female employees.
Alternative Hypothesis (H₁): The mean salary of male employees is significantly higher than that of female employees.
```

```python
#q-8
The p-value is a statistical measure that quantifies the strength of evidence against the null hypothesis in hypothesis testing. It represents the probability of observing the data or more extreme results, assuming that the null hypothesis is true.

In hypothesis testing, the p-value helps determine the level of statistical significance and guides the decision-making process. Here are the key aspects and significance of the p-value:

Interpreting the p-value:

If the p-value is small (typically below the chosen significance level), it suggests strong evidence against the null hypothesis. It indicates that the observed data is unlikely to occur by random chance alone, assuming the null hypothesis is true. This leads to the rejection of the null hypothesis in favor of the alternative hypothesis.
If the p-value is large (greater than the chosen significance level), it suggests weak evidence against the null hypothesis. It indicates that the observed data is reasonably likely to occur by random chance alone, assuming the null hypothesis is true. This leads to the failure to reject the null hypothesis.
Significance Level (α):
The p-value is compared to the significance level (α), which is predetermined before conducting the hypothesis test. Commonly used significance levels are 0.05 (5%) or 0.01 (1%).
If the p-value is smaller than the significance level (p < α), it provides evidence to reject the null hypothesis, indicating a significant result.
If the p-value is greater than the significance level (p ≥ α), it suggests insufficient evidence to reject the null hypothesis, indicating a non-significant result.
Subjective Decision:

The choice of the significance level is subjective and depends on the researcher's willingness to make Type I errors (rejecting a true null hypothesis) and Type II errors (failing to reject a false null hypothesis).
A smaller significance level (e.g., α = 0.01) requires stronger evidence to reject the null hypothesis, reducing the risk of Type I errors but increasing the risk of Type II errors.
A larger significance level (e.g., α = 0.05) allows for weaker evidence to reject the null hypothesis, increasing the risk of Type I errors but reducing the risk of Type II errors
Strength of Evidence:

The p-value does not provide information about the size or practical significance of the observed effect or difference. It only assesses the strength of evidence against the null hypothesis.
A small p-value does not necessarily indicate a large or important effect, while a large p-value does not necessarily imply a lack of practical significance.
```

```python
#q-10
import numpy as np
from scipy import stats

def two_sample_ttest(sample1, sample2):
   
    # Calculate the sample statistics
    n1 = len(sample1)
    n2 = len(sample2)
    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)
    
    # Calculate the pooled standard deviation
    pooled_std = np.sqrt((var1 + var2) / 2)
    
    # Calculate the test statistic
    t_stat = (mean1 - mean2) / (pooled_std * np.sqrt(2 / n1))
    
    # Calculate the degrees of freedom
    df = n1 + n2 - 2
    
    # Calculate the two-tailed p-value
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_stat), df=df))
    
    return t_stat, p_value


# Example usage
sample1 = [82, 85, 88, 90, 79, 92, 86, 88, 85, 90]
sample2 = [75, 78, 82, 80, 83, 77, 84, 79, 81, 85]

t_statistic, p_value = two_sample_ttest(sample1, sample2)

print("T-statistic:", t_statistic)
print("P-value:", p_value)


```

```python
#q-11
The Student's t-distribution, also known as the t-distribution, is a probability distribution that arises in statistical inference when the population standard deviation is unknown. It is used when working with small sample sizes or when the population standard deviation is not known, making it necessary to estimate it from the sample data.

Key characteristics and uses of the t-distribution are as follows:

Shape and Similarity to Normal Distribution:

The t-distribution is symmetric and bell-shaped, similar to the standard normal distribution (z-distribution).
However, the t-distribution has thicker tails compared to the normal distribution, which accommodates the uncertainty introduced by estimating the population standard deviation from limited sample data.
Parameter: Degrees of Freedom (df):

The shape of the t-distribution is determined by the degrees of freedom (df).
The degrees of freedom represent the number of independent observations used to estimate the population parameters. In the t-distribution, it affects the spread and shape of the distribution.
As the degrees of freedom increase, the t-distribution becomes closer to the standard normal distribution.
When to Use the t-Distribution:

The t-distribution is used in situations where the sample size is small (typically less than 30) or when the population standard deviation is unknown.
It is commonly employed in hypothesis testing and constructing confidence intervals for the population mean when working with small samples.
The t-distribution is also applicable when conducting t-tests for comparing two population means or when estimating the difference between two population means.
Hypothesis Testing and Confidence Intervals:

The t-distribution is used in hypothesis testing to determine whether the difference between sample means or the sample mean and the hypothesized population mean is statistically significant.
It is also employed in constructing confidence intervals, which provide a range of plausible values for the population mean.
```

```python
#q-12
The t-statistic is a measure used in hypothesis testing to assess the difference between a sample mean and a hypothesized population mean when the population standard deviation is unknown. It quantifies the difference between the sample mean and the hypothesized mean in terms of the estimated standard error.

The formula for the t-statistic is as follows:

t = (x - μ) / (s / √n)

where:

t is the t-statistic,
x is the sample mean,
μ is the hypothesized population mean,
s is the sample standard deviation, and
n is the sample size.
```
